```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
import os

files = os.listdir('.')
files
```

```python
for fileName in files:
    if fileName[-1] != ")": continue
    file = Path.cwd() / fileName
    with open(file, 'r') as text:
        textfile = text.read()
    #     print(textfile)
    percentages = textfile.split("]")[1].split("^")[0].split("%\n")
    percentages[0] = percentages[0].split("\n")[1]
    percentages = [percentage.split() for percentage in percentages]
    for i in range(len(percentages)-1):
        percentages[i] = int(percentages[i][0])
    percentages = percentages[:-1]

    t = np.arange(len(percentages))
    
    plt.figure()
    plt.plot(t, percentages)
    plt.xlabel("time (second)")
    plt.ylabel("GPU Utilization (%)")
    plt.title(fileName.split("/")[-1].split("(")[-1].split(")")[0])
```
